# Matplotlib
Matplotlib is a powerful plotting library for python that enables the creation of static, animated, and interactive visualisations. It is widely used for data visualisation in data science and analytics.

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2 ,3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title("Basic line plot")
plt.show()
```

```python
x = [1, 2 ,3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, color='red', linestyle="--", marker="o", linewidth=3, markersize=9)
plt.grid(True)
```

```python
x = [1, 2 ,3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 8, 27, 64, 125]

plt.figure(figsize=(9,5))

plt.subplot(1,3,1)
plt.plot(x, y1, color='green')
plt.title('Plot 1')

plt.subplot(1,2,2)
plt.plot(x, y2, color='red')
plt.title('Plot 2')
```

```python
# Bar Plot
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]

plt.bar(categories, values, color='purple')
```

```python
# Histograms

data = [1, 2, 3, 2, 1, 2, 3, 2, 3, 4, 2, 1, 7, 6, 2, 3, 4, 2, 1, 2, 3, 4, 5, 4, 3]

plt.hist(data, bins=5, color="orange")
```

```python
# Scatter plot

x = [1, 2,3 ,4 ,5]
y = [2, 3, 4, 5,6 ]
plt.scatter(x, y, color="blue", marker='x')
```

```python
# Pie Chart

labels = ['A', 'B', 'C', 'D', 'E']
sizes = [30, 20, 40, 10, 35]
colors=['gold', 'yellowgreen', 'orange', 'skyblue', 'pink']
explode = [0.2, 0, 0, 0, 0]

plt.pie(sizes, labels=labels, colors=colors, shadow=True)
```
